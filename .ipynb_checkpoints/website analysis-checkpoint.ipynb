<h1 style="color:#4245f5"> üìà Website Traffic and Engagement Analysis </h1>

### üîç Objective:
Analyze website user traffic, engagement, and channel-wise performance using hourly data from analytics tools (e.g., Google Analytics 4). This report answers key business questions to help optimize user acquisition and content strategy.

### üìÅ Dataset Description:

| Column Name                          | Description |
|-------------------------------------|-------------|
| `Session primary channel group`     | Traffic source (e.g., Direct, Organic Social) |
| `Date + hour (YYYYMMDDHH)`          | Timestamp of traffic in hourly granularity |
| `Users`                             | Unique users in that hour |
| `Sessions`                          | Total sessions started |
| `Engaged sessions`                  | Meaningful sessions (time spent, clicks etc.) |
| `Average engagement time per session` | Time (in seconds) spent actively in each session |
| `Engaged sessions per user`         | Engaged sessions per unique user |
| `Events per session`                | Interactions per session |
| `Engagement rate`                   | % of sessions that were engaged |
| `Event count`                       | Total number of events triggered |


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[0]  # we just replace 1st row as column  and then remove it
df = df.drop(index = 0).reset_index(drop = True) 
# After dropping the row, the original indexes might look like: [1, 2, 3, ...].
#This part resets the index back to [0, 1, 2, ...].  drop=True means: don‚Äôt keep the old index as a new column
df.columns = ["channel group", "DateHour", "Users", "Sessions","Engaged sessions" , "Average engagement time per session" , "Engaged Session per user" , "Events per session" ,	"Engagement rate" ,	"Event count"]
```

```python
df.head()
```

```python
df.info() #Here the datetype are all object even if it's different datatype
```

## Change the datatype as required

```python
# ‚úÖ Step 1: Convert 'DateHour' to datetime
df['DateHour'] = pd.to_datetime(df['DateHour'], format="%Y%m%d%H", errors='coerce') #If you find a value that can‚Äôt be converted, don‚Äôt raise an error ‚Äî just replace it with NaT (for datetime) or NaN (for numeric).

# ‚úÖ Step 2: Select all numeric columns (exclude categorical ones)
numeric_cols = df.columns.drop(['channel group', 'DateHour'])

# ‚úÖ Step 3: Convert those columns to numeric, coercing errors
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df["Hour"] = df['DateHour'].dt.hour
```

```python
df.info()
```

```python
df.isnull().sum()  #There is no null value
```

```python
df.describe()
```

<h3 style="color:#f5e642"> 1) What patterns or trends can you observe in website sessions and users over time?</h3>


```python
sns.set(style = "whitegrid")  #for white grid
```

```python
plt.figure(figsize=(10, 5))

# Group by DateHour and sum Sessions and Users
df.groupby("DateHour")[["Sessions", "Users"]].sum().plot(ax=plt.gca(), color=['crimson', 'royalblue'])

plt.title("Sessions and Users Over Time")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Optional: Rotate for better readability
plt.tight_layout()
plt.show()
```

#### ‚úÖ Conclusion:

Analysis of the time-series data reveals clear and actionable patterns in website traffic:

- **Peak Traffic Windows**:
  - The highest user activity occurs between **3 PM ‚Äì 8 PM** and **12 AM ‚Äì 3 AM**.
  - These hours show consistent spikes in both sessions and users, indicating optimal times for user engagement.

- **Consistent Alignment**:
  - The trends for both sessions and users are highly correlated, suggesting that most users initiate one session per visit.
  - This reflects a stable usage behavior with limited repeat sessions in short time spans.

- **Low Activity Periods**:
  - Early morning hours (especially **6 AM ‚Äì 11 AM**) see a significant drop in traffic, indicating low user availability.

- **Daily Usage Rhythm**:
  - The traffic shows a daily cycle: rising in the afternoon, peaking at night, and dropping by early morning.

#### üìå Business Implication:
To maximize reach and engagement, content publication, ads, and marketing campaigns should be scheduled during the observed **peak engagement windows**. Additionally, low-traffic periods can be explored for background updates or testing new features with minimal disruption.

<h3 style="color:#f5e642">5) Which channels are driving more engaged sessions compared to non-engaged
ones, and what strategies can improve engagement in underperforming
channels?</h3>

```python
df.head()
```

```python
session_df = df.groupby("channel group")[['Sessions' , 'Engaged sessions']].sum().reset_index()
session_df
```

```python
session_df['Non_Engeged'] = session_df['Sessions'] - session_df['Engaged sessions']
session_df_melt = session_df.melt(id_vars = 'channel group',value_vars = ["Engaged sessions","Non_Engeged"])
session_df_melt
```

```python
custom_palette = {
    'Engaged sessions': "#054f84",    # Blue
    'Non_Engeged': "#FFF700"          # Yellow
}

plt.figure(figsize=(10, 5))
sns.barplot(
    data=session_df_melt,
    x='channel group',
    y='value',
    hue='variable',
    palette=custom_palette
)
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

| Channel        | Engaged Sessions | Non-Engaged Sessions | More Engaged? |

| Organic Video  | 109              | 32                   | ‚úÖ Yes         |
| Referral       | 20,653           | 10,337               | ‚úÖ Yes         |
| Unassigned     | 4                | 555                  | ‚ùå No          |


## üîç Insights
‚úÖ High Engagement Channels:

- Organic Social and Referral lead in engaged sessions, showing they attract and retain high-quality traffic.
- Organic Search also performs well, indicating search-optimized content is engaging.
- Organic Video‚Äîthough lower in total sessions‚Äîhas significantly more engaged than non-engaged ones, proving video content‚Äôs deep engagement power.

‚ùå Underperforming Channels:

- Direct traffic shows more non-engaged than engaged sessions, suggesting users coming directly to the site are less engaged.
- Unassigned and Email are low-volume and underperform in engagement.

### ‚úÖ Recommendations
- Boost Video Content Strategy:
 Invest in more video content, especially across platforms driving Organic Video.
- Improve Direct Traffic Experience:
 Optimize homepage UX, CTAs, and load speed to better engage direct visitors.
- Email Campaign Rework:- Test more personalized or visually appealing email campaigns to improve performance.
- Analyze Unassigned Traffic:
 Use better tagging and campaign tracking to understand and improve ‚ÄúUnassigned‚Äù sources.

<h3 style="color:#f5e642">7) Is there any correlation between high traffic (sessions) and high engagement rate over time? </h3>

```python
df.head()
```

```python
correlation = df[['Sessions', 'Engagement rate']].corr()
print(correlation)
```

```python
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.regplot(x='Sessions', y='Engagement rate', data=df , color = 'teal')
plt.title('Correlation Between Sessions and Engagement Rate')
plt.xlabel('Sessions')
plt.ylabel('Engagement Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üìä Insight from the Chart: Correlation Between Sessions and Engagement Rate
This scatter plot with regression line helps visualize the relationship between Sessions and Engagement Rate.

### üîç Key Observations:
#### ‚úÖ Overall Positive Correlation:
As the number of sessions increases, engagement rate tends to increase slightly ‚Äî this is shown by the upward-sloping regression line.

#### ‚ö†Ô∏è High Variability at Low Sessions:

When sessions are low (under 50), engagement rate is highly scattered (ranging from 0 to 1).

This indicates inconsistency ‚Äî some users are highly engaged even with low traffic.

#### üìà More Stable Engagement at High Sessions:

As sessions exceed 100+, engagement rate appears to stabilize and concentrate around mid-to-high values (0.4 to 0.7).

This suggests more reliable engagement with higher traffic.

#### üßØ Outliers:

A few points have very high engagement at low sessions and very low engagement even at high sessions ‚Äî these might be anomalies or specific behaviors (like bot traffic or abandoned sessions)

```python
df_plot = df.groupby("DateHour")[["Engagement rate","Sessions"]].mean().reset_index()

plt.figure(figsize=(12,5))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color = "red")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="lightseagreen")
plt.title("Engagement Rate by Binned Session Count")
plt.legend()
plt.grid(True)
plt.show()
```
